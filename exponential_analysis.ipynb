```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.signal import find_peaks
import time
import sounddevice as sd
import noisereduce as nr


class LogSpectrum:
    def __init__(self):
        self.sample_rate = 44100 # sampling rate
        self.Noctave = 24 # octave resolution
        self.Dt = 1/self.sample_rate            # sampling time step
        self.omega0 = 10  # Hz, the smallest frequency
        self.krange = range(10*self.Noctave)
        self.omega = np.array([ 2**(k/self.Noctave)*self.omega0 for k in self.krange])
        self.tint = 0.1 #sec
        self.Nint = int(self.tint*self.sample_rate)
        self.K = np.empty((len(self.krange),self.Nint))
        self.L = np.empty((len(self.krange),self.Nint))
        self.spectrum_data = None

    def set(self, data):
        self.data = data
        self.maxtime = len(data)/self.sample_rate
        return self

    def update(self):
        self.spectrum_data = None
        rfactor = 2**(1/self.Noctave)-1 
        for i,k in enumerate(self.krange):
            oo = self.omega[k]
            f = np.array([ np.exp(-(rfactor*oo*n*self.Dt)**2/2) for n in range(self.Nint) ])
            self.K[i] = np.array([ f[n]*np.sin(2*np.pi*oo*n*self.Dt)*oo for n in range(self.Nint) ])
            self.L[i] = np.array([ f[n]*np.cos(2*np.pi*oo*n*self.Dt)*oo for n in range(self.Nint) ])
        return self

    def convert_range(self, t1=0, t2=None):
        n1 = int(t1*self.sample_rate)
        if t2 is not None:
            n2 = int(t2*self.sample_rate)
        else:
            n2 = len(self.data)
        return n1,n2

    def spectrum(self, t=0):
        if t<0 or t+self.tint > self.maxtime:
            print(f'time {t} is out of range: it must be in [0,{self.maxtime}]')
        self.starttime = t
        n,_ = self.convert_range(t)
        self.spectrum_data = (self.K@self.data[n:n+self.Nint]/self.Nint)**2 + (self.L@self.data[n:n+self.Nint]/self.Nint)**2
        self.maxvalue = np.max(self.spectrum_data)
        self.spectrum_data /= (1e-5 + self.maxvalue)
        return self.spectrum_data

    def play(self,t1 = 0, t2=None):
        n1,n2 = self.convert_range(t1,t2)
        channel_0_int16 = np.int16(self.data[n1:n2] / np.max(np.abs(self.data)) * 32767)
        sd.play(channel_0_int16, samplerate=self.sample_rate)
        sd.wait()

    def plot_data(self, ax = None, **kwargs):
        if ax is None:
            fig, ax = plt.subplots()
        n1,n2 = self.convert_range(self.starttime, self.starttime+self.tint)
        subt = np.arange(len(self.data[n1:n2]))/self.sample_rate + self.starttime
        ax.plot(subt, self.data[n1:n2], **kwargs)
        return ax
    
    def plot_spectrum(self, ax = None, **kwargs):
        if self.spectrum_data is None:
            self.spectrum()
        if ax is None:
            fig, ax = plt.subplots()
        ax.plot(self.omega, self.spectrum_data, **kwargs)
        ax.set_xscale('log')
        return ax

    def max_freq(self, map = (lambda x: np.log(x), lambda x:np.exp(x))):
        if self.spectrum_data is None:
            self.spectrum()
        index = np.argmax(self.spectrum_data)
        if index == 0 or index ==len(self.spectrum_data)-1:
            return index, self.spectrum_data[index]
        omega_list = map[0](self.omega[index-1:index+2])
        Vlist = self.spectrum_data[index-1:index+2]
        T = np.array([omega_list**2, omega_list, np.ones(omega_list.shape)]).T
        coeff = np.linalg.inv(T)@Vlist
        return map[1](-coeff[1]/(2*coeff[0]))

    def plot_octaves_2D(self, ax = None, **kwargs):
        if self.spectrum_data is None:
            self.spectrum()
        if ax is None:
            fig, ax = plt.subplots()
        res= self.spectrum_data.copy()
        res = res.reshape(-1,self.Noctave)
        x = np.arange(res.shape[1]+1)
        y = np.arange(res.shape[0]+1)    
        mesh = ax.pcolormesh(x, y, res, **kwargs)
        return mesh
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd

# Sampling rate and duration
fs = 44100
duration = 2  # seconds
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Base motor frequency (like RPM / 60)
f_base = 60  # 60 Hz = 3600 RPM

# Create harmonics (simulate engine buzz)
motor = np.zeros_like(t)
for i in range(1, 10):
    motor += np.sin(2 * np.pi * f_base * i * t) * (1 / i)  # add harmonics

# Apply amplitude modulation (like revving)
#modulation = 0.5 * (1 + np.sin(2 * np.pi * 2 * t))  # 2 Hz modulation
#motor *= modulation

# Normalize for playback
motor /= np.max(np.abs(motor))

# Listen
sd.play(motor, fs)
sd.wait()

# Optional: visualize
plt.plot(t[:1000], motor[:1000])  # zoom in on waveform
plt.title("Simulated Motor Sound")
plt.show()
```

```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt

fs = 44100
duration = 2.5  # seconds
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Simulate 4 rotors spinning at different base frequencies
rotor_freqs = [230, 245, 260, 275]  # Hz

drone_sound = sum(np.sin(2 * np.pi * f * t) for f in rotor_freqs)

# Add amplitude modulation for flutter
mod = 0.7 + 0.3 * np.sin(2 * np.pi * 10 * t)  # 10 Hz flutter
drone_sound *= mod

# Add filtered noise (blade hiss)
noise = 0.2 * np.random.randn(len(t))
# Apply a simple low-pass filter
from scipy.signal import butter, lfilter
b, a = butter(6, 0.2)
filtered_noise = lfilter(b, a, noise)

# Mix it in
drone_sound += filtered_noise

# Normalize
drone_sound /= np.max(np.abs(drone_sound))

# Listen
sd.play(drone_sound, fs)
sd.wait()

# Optional: Plot a small snippet
plt.plot(t[:1000], drone_sound[:1000])
plt.title("Simulated Drone Sound")
plt.show()
```
